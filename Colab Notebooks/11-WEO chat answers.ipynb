## Execute code from here so no need to re-index information

```python
import os
import openai
from llama_index import VectorStoreIndex, SimpleDirectoryReader

openai.api_key = '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'

# load index from persisting file
from llama_index import StorageContext, load_index_from_storage

# rebuild storage context
storage_context = StorageContext.from_defaults(persist_dir="/content/drive/MyDrive/PHB009/dataset/News/WEO/")
# load index
index2 = load_index_from_storage(storage_context)

query_engine = index2.as_query_engine()
```

```python
# Chat start
while True:
    user_input = input('How can I help? ')
    if user_input.lower() == 'q':
        print('Exiting...')
        break
    response = query_engine.query(user_input)
    print(f'Response: {response}\n')
```
